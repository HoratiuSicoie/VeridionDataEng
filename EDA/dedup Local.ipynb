```python
import pandas as pd
df = pd.read_csv("../Data/cleaned_products.csv")

df.head()
```

```python
from deduplipy.blocking import Blocking
from deduplipy.deduplicator import Deduplicator
```

```python
#CHOOSE THE BEST COLUMNS FOR DEDUP:
for col in df.columns:
    print(f"{col} - unique: {df[col].nunique()} - missing: {df[col].isna().sum()}")

#Look for uniqueness/missingness trade-off
#Low uniqueness → possibly not helpful
#High missingness → too sparse to rely on

```

```python
#Based on ABOVE:
# USE AS INITIAL['product_title', 'product_name', 'brand'] 

#ADD and evaluate further on : ['description','product_summary']
```

```python
dedup_configs = {
    "Model 1": ['product_title', 'product_name', 'brand']}
#here possibility to run out of memory when ran locally with other attribute included such as description so stick with this for now
#for local run stick to this for now.
```

```python
from deduplipy.deduplicator import Deduplicator

dedup_columns = dedup_configs["Model 1"] 
for col in dedup_columns:
    df[col] = df[col].fillna('').astype(str)

deduper = Deduplicator(
    col_names=dedup_columns,
    interaction=True,
    verbose=1
)
deduper.fit(df)
```

```python
df_deduped = deduper.predict(df, score_threshold=0.2, cluster_threshold=0.6)
```
